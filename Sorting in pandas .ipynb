# Sorting operation in python

```python
import pandas as pd

# Sample data dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, None, 14],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [50000, None, 70000, 80000],
    'Perfomance Score': [78, None, 90, 80]
}

# Convert the dictionary into two DataFrames
df1 = pd.DataFrame(data)
df1
```

# Sorting 

```python
# df1.sort_value(by='column_name ', True and False ascending and discending ,inplace = True for change the original data)
```

```python
df1.sort_values(by='Age', ascending= False)
df1
```

```python
df1.sort_values(by='Age', ascending =True)
df1
```

```python
df1.sort_values(by='Salary', ascending=False)
# for descending order 
```

# Sorting by Multiple Columns 

```python
df1.sort_values(by=['City', 'Age'], ascending=[True, False])
```
