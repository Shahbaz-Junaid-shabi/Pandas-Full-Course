# fillna(calue_default or set value , inplace = True ) method 

```python
import pandas as pd

# Sample data dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, None, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [50000, None, 70000, 80000],
    'Perfomance Score': [78, None, 90, 80]
}

# Convert the dictionary into two DataFrames
df1 = pd.DataFrame(data)
df1
```

```python
df1.fillna(0,inplace = True)
df1
```

# If we you want to fill calculated value 

```python
df1['Age'].fillna(df1['Age'].mean(),inplace = True)
df1['Salary'].fillna(df1['Salary'].mean(),inplace = True)
df1
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
df1['Salary'] = df1['Salary'].fillna(df1['Salary'].mean())
df1
```
