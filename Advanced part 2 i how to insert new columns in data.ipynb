```python
import pandas as pd

# Sample data dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [50000, 60000, 70000, 80000],
    'Perfomance Score': [78, 50, 90, 80]
}

# Convert the dictionary into two DataFrames
df1 = pd.DataFrame(data)
df1
```

```python
df1["Bonus"] = df1['Salary'] * 0.1 #add new column and increase with 10%
df1
```

```python
df1["Eid Bonus"] = df1['Salary'] + 3000
df1
```

```python
df1['Increament Salary'] = df1['Salary'] * 0.02
df1
```

# Using INSERT Method to insert the new column in data 

```python
# using the INSERT method we insert column on specific location where we want and also as per requirement

# df1.insert(loc,column_name,value)
df1.insert(0,'Empolyee ID',range(1,len(df1) + 1))
df1
```

```python
# Delete the 'Employee ID' column from the DataFrame
df1.drop('Empolyee ID', axis=1, inplace=True)

# Display the updated DataFrame
df1
```
